```python
import re

text="Iam Teja Swaroop My email is teja9748@gmail.com sai787@email.com"
email_templ=re.compile('[a-zA-Z0-9_.]+@[a-zA-Z0-9]+[.]+[a-zA-Z]{2,4}')

l=email_templ.findall(text)
for email in l:
  print(email)
```

```python
paswd=input('Enter The password:')

def valid_passwd(paswd):
  if not len(paswd)>5:
    return False
  elif not re.search('[a-z]',paswd):
    return False
  elif not re.search('[A-Z]',paswd):
    return False
  elif not re.search('[0-9]',paswd):
    return False
  elif not re.search('[#@$_ ]',paswd):
    return False
  return True

if(valid_passwd(paswd)):
  print("valid")
else:
  print("invalid")
```

```python
import pandas as pd

data=pd.read_csv("/content/drive/MyDrive/datasets/FremontBridge.csv",index_col='Date',parse_dates=True)

data
```

```python
# i) Average daily bicycle counts
avg_daily_count=data.groupby(data.index.dayofweek).mean()
print(avg_daily_count)

avg_daily_count.index=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
avg_daily_count.plot(style=[':','-','--'])
```

```python
# ii) Average hourly bicycle counts by weekday and weekend.
import numpy as np
days=np.where(data.index.dayofweek <5,'Weekday','Weekend')
print(days)
avg_hourly_count_data=data.groupby([days,data.index.time]).mean()
print(avg_hourly_count_data)
avg_hourly_count_data.loc['Weekday'].plot(title='Weekdays',style=[':','--','*'])
avg_hourly_count_data.loc['Weekend'].plot(title='Weekends',style=[':','--','*'])


```

```python
df=pd.read_csv('/content/drive/MyDrive/datasets/births.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
# i) Total number of US births by year and gender
births_by_year=df.groupby(['year','gender'])['births'].sum()
print(births_by_year,'\n')
total=df.pivot_table('births',index='year',columns='gender',aggfunc='sum')

total

```

```python
df['day'].unique()
```

```python
df=df[(df['day'] >0) & (df['day']<31)]
df
```

```python
df.isnull().sum()
```

```python
df['date']=pd.to_datetime(df[['year','month','day']],format='%y%m%d',errors='coerce')

df['day_of_week_num']=df['date'].dt.day_of_week

df['decade']=(df['year'] //10)*10
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# ii) Average daily births by day of week and decade
avg_daily_births=df.groupby(['decade','day_of_week_num'])['births'].mean()
print(avg_daily_births)
df.pivot_table('births',index='day',columns='decade',aggfunc='mean').plot()
```
